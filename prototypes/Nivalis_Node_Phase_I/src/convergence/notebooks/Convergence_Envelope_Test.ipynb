# Halessa Phase II — Convergence Envelope Field Sweep

This notebook sweeps the fusion field model defined in `fusion_field.py`, visualizing the spatial behavior of the convergence envelope Q(x, y, z, t) over a 2D plane at fixed z, t.

```python
import numpy as np
import matplotlib.pyplot as plt
from fusion_field import Q_total
```

## Grid Setup

```python
# Define spatial grid
x = np.linspace(-2, 2, 300)
y = np.linspace(-2, 2, 300)
X, Y = np.meshgrid(x, y)

# Set fixed z and t values
z = 0.0
t = 0.0
```

## Compute Q(x, y) field

```python
# Evaluate field
Q_field = Q_total(X, Y, z, t)
```

## Visualize

```python
plt.figure(figsize=(10, 8))
plt.contourf(X, Y, Q_field, 100, cmap='plasma')
plt.colorbar(label='Q(x, y) Convergence Intensity')
plt.title("Halessa Phase II: Convergence Envelope Field at z=0, t=0")
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')
plt.grid(False)
plt.tight_layout()
plt.show()
```
